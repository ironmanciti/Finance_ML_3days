# 두 random 변수간의 연관성 - Correlation Coefficient (상관계수)

## covariance 

- 한 변수의 variance 가 얼마나 많이 다른 변수의 variance 로 설명되는지를 측정한 것. 즉, Y 가 변할 때마다, X 가 얼마나 변하는지 check 하여 따라서 변화하는 정도가 크면 high covariance 를 갖는다고 말함.  

- Covariance 자체로는 큰 의미가 없다.


## 상관계수

- 두개의 변수가 linear 한 정도를 측정하는 것  

- Correlation 은 단순히 정규화된 형태의 covariance 이다. 정규화하지 않으면 correlation 과 covariance 는 일상 대화에서 거의 비슷한 의미로 사용된다. 두개는 개념적으로는 거의 동일하다. 


- 두 변수는 price 와 price 가 될 수도 있고, price 와 강우량이 될 수도 있다. 어떤 조합도 가능  

- 그 값은 언제나 -1 과 1 사이  

- 양의 계수는 두개의 변수가 직접 연결된 것을 표시. 하나가 증가하면 다른 하나도 증가.  

- 음의 계수는 두개의 변수가 역으로 관련된 것을 표시. 하나가 증가하면 다른 하나는 감소.   

- 상관계수가 0 에 가까워질수록 두 변수간의 관계는 약화

- 두개의 random data set 은 0 에 가까운 correlation 을 갖는다.

- 두 변수 $X$ 와 $Y$ 가 다음과 같이 정의되면, 상관계수는
$$r = \frac{Cov(X,Y)}{std(X)std(Y)}$$

```python
import numpy as np
import matplotlib.pyplot as plt
import FinanceDataReader as fdr
import pandas as pd
pd.set_option('display.max_rows', None)

import yfinance as yf
from pykrx import stock
import datetime
import korean

# 투자 기간을 1년으로 설정
num_of_years = 1

# 오늘 날짜로부터 1년 전의 날짜를 계산합니다. 365.25를 곱하여 윤년을 고려합니다.
start = datetime.date.today() - datetime.timedelta(days=int(365.25 * num_of_years))
# 오늘 날짜를 종료 날짜로 설정
end = datetime.date.today()
start, end
```

## 두개의 random variable 의 correlation

- 서로간의 correlation이 없는 두 개의 변수

```python
X = np.random.rand(50)
Y = np.random.rand(50)

corr_XY = np.corrcoef(X, Y)
print(corr_XY)
```

- correlation 을 부여한 두개의 random variable을 비교

```python
X = np.random.rand(50)
Y = X + np.random.normal(0, 0.1, 50)       # 약간의 noise 추가

print(np.corrcoef(X, Y))
print()

plt.scatter(X,Y)
plt.xlabel('X Value')
plt.ylabel('Y Value')

print('Correlation 직접 계산: ' + str(np.cov(X,Y)[0, 1] / (np.std(X)*np.std(Y))))
print('numpy 내장 함수 사용: ' + str(np.corrcoef(X, Y)[0, 1]))
```

역의 상관 관계를 만들어 본다.

```python
X = np.random.rand(50)
Y = -X + np.random.normal(0, .1, 50)

print(np.corrcoef(X, Y))
print()

plt.scatter(X,Y)
plt.xlabel('X Value')
plt.ylabel('Y Value')

print('Correlation: ' + str(np.corrcoef(X, Y)[0, 1]))
```

## Correlation 의 금융 data 적용

### 관련된 자산들을 파악

- 두개의 자산이 correlated 라는 것은 움직임이 비슷하다는 의미  

- 주의할 점은 두개의 자산이 correlated 라는 것은 움직임이 비슷하다는 것이지 한 자산의 움직임이 다른 자산의 움직임을 일으킨다는 의미는 아니란 것이다.

### 서로 연관성 없는 자산들로 portfolio 구성

finance 에서 correlation 이 유용한 이유는 portfolio 구성에 있다. 연관성 없는 자산들로 portfolio 를 구성하면 안정된 수익을 기대할 수 있다.

### 상관계수의 한계점

- 변수들이 정규분포 되어 있지 않은 경우 특히 correlation 이 중요한지 말하기 쉽지 않다. 상관계수는 1 에 가까운 경우 측정 기간 동안 두개의 주식이 correlate  되어있다고 말할 수 있지만 미래에도 그럴지 알 수 없다.   
- KOSPI 지수와 correlation 이 높은 주식들은 다른 평균적인 주식 보다 개별 주식간의 correlation 이 좀 더 높을 것이다.  
- 특정 시점에만 correlated 된 것 처럼 보일 수 있으므로 긴 기간에 걸쳐 상관계수를 검사해야 한다. 긴 기간 동안 correlation 이 얼마나 stable 한가 측정할 필요가 있다.

### 다양한 주식 종목 간의 correlation 파악

```python
# 티커와 한글 종목명을 매핑하는 딕셔너리
ticker_to_name = {
    '005930': '삼성전자',
    '007630': '포스코퓨처엠',
    '009150': '삼성전기',
    '000660': 'SK하이닉스',
    '051910': 'LG화학',
    '035420': 'NAVER',
    '006400': '삼성SDI',
    '068270': '셀트리온',
    '005380': '현대차',
    '207940': '삼성바이오로직스',
    '035720': '카카오',
    '012330': '현대모비스',
    '096770': 'SK이노베이션',
    '055550': '신한지주',
    '032830': '삼성생명'
}

symbols_list = ticker_to_name.keys()
symbols_list
```

```python
# 조회를 시작할 날짜
start = datetime.datetime(2022, 1, 1)

# 주식 가격 데이터를 저장할 배열입니다.
symbols_prices = []

for ticker in symbols_list:
    r = fdr.DataReader(ticker, start)
    # 가져온 데이터가 비어있지 않은 경우에만 다음 단계를 진행
    if not r.empty:
        # 가져온 데이터에 'Symbol' 컬럼을 추가하고, 티커 이름을 값으로 설정합니다.
        r['Symbol'] = ticker
        symbols_prices.append(r)
    else:
        # 데이터가 비어 있는 경우, 해당 티커를 스킵하고 콘솔에 메시지를 출력합니다.
        print(f"No data found for {ticker}")

symbols_prices
```

```python
# 여러 데이터프레임이 저장된 symbols 리스트를 하나의 데이터프레임으로 합칩니다.
df = pd.concat(symbols_prices)

# 인덱스를 재설정하여 합친 데이터프레임의 인덱스가 연속적이고 정렬되도록 합니다.
df = df.reset_index()

# 필요한 컬럼만을 선택하여 데이터프레임을 재구성합니다.
df = df[['Date', 'Close', 'Symbol']]
df.head()
```

```python
# 'Date'를 인덱스로, 'Symbol'을 컬럼으로, 'Close' 값을 value로 사용하여 피벗 테이블을 생성
df_pivot = df.pivot(index='Date', columns='Symbol', values='Close')

df_pivot.head()
```

```python
# 피벗 테이블의 컬럼명을 한글명으로 변경
new_columns = [ticker_to_name.get(item, item) for item in df_pivot.columns]

df_pivot.columns = new_columns

df_pivot.head()
```

```python
# df_pivot 데이터프레임의 컬럼 간 상관계수 계산
corr_df = df_pivot.corr()
corr_df
```

```python
import seaborn as sns

plt.figure(figsize=(13, 8))
# annot=True는 각 셀에 상관계수 값을 표시하라는 옵션이며, cmap="RdYlGn"은 색상 맵을 지정
# "RdYlGn"은 상관계수의 강도에 따라 빨간색에서 녹색으로 색상 변화
sns.heatmap(corr_df, annot=True, cmap="RdYlGn")
plt.show()
```
