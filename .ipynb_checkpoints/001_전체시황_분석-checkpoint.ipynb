## 3. 시장 매매 주체 현황 

```python
delta = 90
try:
    now = datetime.datetime.now()
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 4))

    df = stock.get_market_trading_value_by_date(now-datetime.timedelta(delta), now, "KOSPI", freq="m")
    df.drop(['기타법인', '전체'], axis=1, inplace=True)
    df.plot(kind='bar', ax=ax1)
    ax1.set_title(f"최근 {delta}일 간 매매 주체별 매수/매도 현황 - KOSPI")
    ax1.set_xticklabels(df.index, rotation=45)

    df = stock.get_market_trading_value_by_date(now-datetime.timedelta(delta), now, "KOSDAQ", freq="m")
    df.drop(['기타법인', '전체'], axis=1, inplace=True)
    df.plot(kind='bar', ax=ax2)
    ax2.set_title(f"최근 {delta}일 간 매매 주체별 매수/매도 현황 - KOSDAQ")
    ax2.set_xticklabels(df.index, rotation=45)
    plt.show()
except:
    pass
```

### KOSPI 매수 누계

```python
delta = 500
try:
    daily = stock.get_market_trading_value_by_date(now-datetime.timedelta(delta), now, "KOSPI")
    daily_kospi = daily.copy()
    daily['기관누계'] = daily['기관합계'].cumsum()
    daily['외국인누계'] = daily['외국인합계'].cumsum()
    daily['개인누계'] = daily['개인'].cumsum()
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(20, 16))
    daily['외국인누계'].plot(ax=ax1, title="KOSPI 외국인누계")
    daily['기관누계'].plot(ax=ax2, title="KOSPI 기관누계")
    daily['개인누계'].plot(ax=ax3, title="KOSPI 개인누계")
    plt.tight_layout()
    
    display(HTML(daily.resample('M').sum().tail().to_html()))
except Exception as e:
    print(e)
```

### KOSDAQ 매수 누계

```python
try:
    daily = stock.get_market_trading_value_by_date(now-datetime.timedelta(delta), now, "KOSDAQ")
    daily_kosdaq = daily.copy()
    daily['기관누계'] = daily['기관합계'].cumsum()
    daily['외국인누계'] = daily['외국인합계'].cumsum()
    daily['개인누계'] = daily['개인'].cumsum()
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(20, 16))
    daily['외국인누계'].plot(ax=ax1, title="KOSDAQ 외국인누계")
    daily['기관누계'].plot(ax=ax2, title="KOSDAQ 기관누계")
    daily['개인누계'].plot(ax=ax3, title="KOSDAQ 개인누계")
    plt.tight_layout()
    
    display(HTML(daily.resample('M').sum().tail()))
except Exception as e:
    print(e)
```

### KOSPI+KOSDAQ 한국 시장 전체 매수 누계

```python
try:
    daily_all = pd.DataFrame()
    daily_all['기관합계'] = daily_kospi['기관합계'] + daily_kosdaq['기관합계']
    daily_all['개인'] = daily_kospi['개인'] + daily_kosdaq['개인']
    daily_all['외국인합계'] = daily_kospi['외국인합계'] + daily_kosdaq['외국인합계']
    daily_all['기관누계'] = daily_all['기관합계'].cumsum()
    daily_all['외국인누계'] = daily_all['외국인합계'].cumsum()
    daily_all['개인누계'] = daily_all['개인'].cumsum()
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(20, 16))
    daily_all['외국인누계'].plot(ax=ax1, title="전체 외국인누계")
    daily_all['기관누계'].plot(ax=ax2, title="전체 기관누계")
    daily_all['개인누계'].plot(ax=ax3, title="전체 개인누계")
    plt.tight_layout()
    display(HTML(daily_all.resample('M').sum().tail()))
except Exception as e:
    print(e)
```

## 4. 연속 상승/하락 pattern

```python
start_1 = "2023-02-01"
try:
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 8))
    kospi = yf.download('^KS11', start=start_1, end=now, progress=False)
    ax1.plot(kospi['Close'])
    ax1.grid()
    ax1.margins(x=0)
    ax1.set_title("KOSPI")

    kospi = kospi.drop(['Open', 'High', 'Low', 'Adj Close', 'Volume'], axis=1)
    kospi = kospi.diff()
    kospi['color'] = ['red' if kospi.loc[idx, 'Close'] > 0 else 'blue' for idx in kospi.index]
    color_list = kospi['color'].values.tolist()
    ax2.bar(kospi.index, kospi['Close'], color=color_list)
    kospi['Close'].plot(kind='bar', color=color_list, figsize=(20, 4))
    ax2.set_xticks([])
    ax2.grid()
    ax2.margins(x=0)
    plt.show()
    
    print(f'{start2.strftime("%Y%m%d")} - {now.strftime("%Y%m%d")} 기간 중')
    print(Counter(kospi['color'].values))
except:
    pass
```
