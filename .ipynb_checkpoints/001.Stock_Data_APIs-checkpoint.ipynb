# Free stock market APIs

주식 시장 데이터 API는 현재 시장에서 거래되는 금융 자산에 대한 실시간 또는 과거 데이터를 제공합니다. 이러한 API는 일반적으로 공개 주식, ETF, ETN의 가격을 제공합니다.  
이러한 데이터는 거래 전략을 수립하고 시장을 모니터링하는 기반이 되는 기술 지표를 생성하는 데 사용할 수 있습니다.

```python
import matplotlib.pyplot as plt
import yfinance as yf
import datetime
```

### Yahoo Finance 사용 방법 1:

- yf.Ticker 클래스를 사용하여 특정 주식에 대한 객체를 생성  
- history() 메서드를 호출하여 주식의 과거 데이터를 검색  
- history 외에도 주식의 배당금(dividends), 액면분할(splits), 재무제표(financials), 주주 정보(institutional_holders) 등 추가적인 정보에 쉽게 접근 가능

```python
goog = yf.Ticker('goog')
data = goog.history()
data.head()
```

### Yahoo Finance 사용 방법 2:
- yf.download 함수를 사용하여 지정된 기간 동안의 주식 데이터를 바로 다운로드
- 데이터를 빠르고 간단하게 조회할 때 유용

```python
# 단일 주식 주가 조회
data = yf.download('GOOG', start='2021-12-10', end='2021-12-30')
data.head()
```

```python
# 여러 주식 주가 동시 조회
data = yf.download(['GOOG','META'], start='2021-12-10', end='2021-12-30', group_by='ticker')
data.head()
```

### 실습

Microsoft의 2024.1.1 ~ 2024.1.31 사이의 주가 조회

```python
import yfinance as yf
import datetime

# Microsoft (MSFT) 주식을 가져올 기업을 생성합니다.
ticker = yf.Ticker("MSFT")

# 시작 날짜와 종료 날짜를 정의합니다.
start_date = datetime.datetime(2023, 1, 1)
end_date = datetime.datetime(2023, 1, 31)

# 특정 시작 날짜와 종료 날짜로 ticker 주식의 역사적 데이터를 가져옵니다.
hist = ticker.history(start=start_date, end=end_date)
hist.head()
```

```python
df = yf.download('MSFT', start=start_date, end=end_date, progress=False)
df.tail()
```

### 한국 주식 주가 조회

```python
df = yf.download('005930.KS', start='2022-01-01', progress=False) #삼성전자
df.tail()
```

### PyKrx

```python
#!pip install pykrx
```

```python
from pykrx import stock
from pykrx import bond
```

- 코스피 시장에 상장된 ticker

```python
tickers = stock.get_market_ticker_list(market="KOSPI")
print(len(tickers))
tickers = stock.get_market_ticker_list(market="KOSDAQ")
print(len(tickers))
tickers = stock.get_market_ticker_list(market="KONEX")
print(len(tickers))
tickers = stock.get_market_ticker_list(market="ALL")
print(len(tickers))
```

- 종목명 반환

```python
for ticker in stock.get_market_ticker_list():
    종목 = stock.get_market_ticker_name(ticker)
    print(종목)
    break
```
