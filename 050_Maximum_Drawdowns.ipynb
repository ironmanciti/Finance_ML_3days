# Maximum Drawdown 계산 

파이썬 프로그램을 이용하여 고점 대비 최대 손실 폭(MDD, Maximum Drawdown)을 구해 보겠습니다.  
우선 필요한 라이브러리를  임포트 합니다. 테슬라의 주가를 이용하여 MDD를 계산해 봅니다.

```python
import numpy as np
import pandas as pd
import datetime
import yfinance as yf
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (4, 3)

start = datetime.datetime(2016, 1, 1)

tesla = yf.download('TSLA', start=start, progress=False)
tesla.head(3)
```

```python
tesla = tesla[['Close']]
```

## Drawdown 계산

step 1. 이전 피크 계산  
step 2. 이전 피크와 현재 값의 차이(비율)로 드로다운을 계산합니다.  

step 1. 이전 피크 계산 - cummax() 메소드를 이용하여 계산합니다.

```python
tesla['previous_peaks'] = tesla['Close'].cummax()
tesla.head()
```

```python
tesla['previous_peaks'].plot()
```

step 2. 이전 피크와 현재 값의 차이(비율)로 드로다운을 계산합니다.  

```python
drawdown = (tesla['Close'] - tesla['previous_peaks']) / tesla['previous_peaks']
drawdown.plot()
```

최대 드로다운을 구합니다.

```python
drawdown.min()
```

최대 드로다운이 발생한 일자를 구합니다.

```python
drawdown.idxmin()
```

테슬라 주식의 가격과 이전 고점 가격을 시각화 합니다.

```python
plt.plot(tesla)
plt.show()
```

Tesla 주식의 역사적 MDD 와 발생 시점은 다음과 같습니다.

```python
print(f'최대 MDD 비율 : {drawdown.min() * 100:.1f} %')
print(f'최대 MDD 발생 일자 : {drawdown.idxmin().date()}')

idmax = tesla['previous_peaks'].idxmax()
print(f'이전 고점의 주식 가격 : US$ {tesla.loc[idmax, "Close"]:.2f}')
print(f'최대 MDD 발생시의 주식 가격 : US$ {tesla.loc[drawdown.idxmin(), "Close"]:.2f}')
```
